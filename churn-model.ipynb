```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
data['SeniorCitizen'] = data['SeniorCitizen'].astype(object)
data.TotalCharges = pd.to_numeric(data.TotalCharges, errors='coerce')
```

```python
data.dropna(how = 'any', inplace = True)
data.drop(axis=1, columns= 'customerID', inplace=True)
```

```python
print(data[data.Churn == 'No']['Churn'].value_counts() *100/ len(data))
print(data[data.Churn == 'Yes']['Churn'].value_counts() *100/ len(data))
```

```python
columns= ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn', 'Contract', 'PaymentMethod']
```

```python
df=pd.get_dummies(data, columns= columns)
```

```python
df
```

```python
df.columns
```

```python
drop_col= ['gender_Female', 'SeniorCitizen_0', 'Partner_No', 'Dependents_No', 
          'PhoneService_No', 'MultipleLines_No phone service', 'InternetService_No',
          'OnlineSecurity_No internet service', 'OnlineBackup_No internet service', 
           'DeviceProtection_No internet service', 'TechSupport_No internet service',
          'StreamingTV_No internet service', 'StreamingMovies_No internet service',
          'PaperlessBilling_No', 'Churn_No', 'Contract_Two year', 'PaymentMethod_Mailed check']
```

```python
df.drop(axis=1, columns= drop_col, inplace=True)
```

```python
df.shape
```

```python
x=df.drop(axis=1, columns= ['Churn_Yes'])
```

```python
y=df.Churn_Yes
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

# **DecisionTreeClassifier**

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
y_predict = model_dt.predict(x_test)
model_dt_score = model_dt.score(x_test, y_test)
print(model_dt_score)
print(metrics.classification_report(y_test, y_predict))
```

# **RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_rf.fit(x_train,y_train)
y_predict = model_rf.predict(x_test)
model_rf_score = model_rf.score(x_test, y_test)
print(model_rf_score)
print(metrics.classification_report(y_test, y_predict))
```

# **LogisticRegression**

```python
from sklearn import linear_model
```

```python
model_logr = linear_model.LogisticRegression(max_iter= 1000)
model_logr.fit(x_train,y_train)
y_predict = model_logr.predict(x_test)
model_logr_score = model_logr.score(x_test, y_test)
print(model_logr_score)
print(metrics.classification_report(y_test, y_predict))
```

# **SupportVectorClassifier**

```python
from sklearn.svm import SVC  
```

```python
model_svc = SVC(kernel='linear') 
model_svc.fit(x_train,y_train)
y_predict = model_svc.predict(x_test)
model_svc_score = model_svc.score(x_test, y_test)
print(model_svc_score)
print(metrics.classification_report(y_test, y_predict))
```
